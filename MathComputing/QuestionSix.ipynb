```python
import numpy as np
import matplotlib.pyplot as plt

number_of_customers = 5
number_of_random_samples = 1000


#generating  1000 random samples
samples= np.random.poisson ( number_of_customers, number_of_random_samples) 

```

```python
#calculate and print the mean and variance of the generated samples 
mean = np.mean(samples)
varience = np.var(samples)

print (f"Mean :{mean}")
print (f"Varience :{varience}")

```

```python
#plot a histogram of the samples to visualise the distribution

plt.figure(figsize=(8,6))
plt.hist(samples, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.title ("Histogram of Customer Samples to Visualise Distribution", fontsize= 14)
plt.xlabel("Number of Customers", fontsize= 10)
plt.ylabel("Frequency", fontsize= 10)
plt.grid(axis="y" , linestyle="--", alpha=0.7)
plt.show()
```
